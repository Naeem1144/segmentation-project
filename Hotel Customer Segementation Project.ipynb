## Outliers Detection and transformation

```python
df.Age.max()
```

```python
df.Age.min()
```

```python
def outliers(x):
    print(x)
    q1 = np.percentile(df[x],25)
    q3 = np.percentile(df[x],75)
    iqr = q3 - q1
    sp = q1 - (1.5*iqr)
    ep = q3 + (1.5*iqr)
    global new_df
    new_df = df[x][df[x].between(sp,ep)]
    print("old shape :",df[x].shape)
    print("new shape :",new_df.shape)
    print("old lower limit :",df[x].min())
    print("old upper limit :",df[x].max())
    print("new lower limit :",new_df.min())
    print("new upper limit :",new_df.max())
    print("Note : Access transformed series object from \'new_df\' variable")
    print()
```

```python
outliers('Age')
```

```python
df['Age'] = new_df[new_df >18]
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
for i in df.select_dtypes('number'):
    outliers(i)
```

## KMEANS

```python
km = KMeans()
```

```python
def elbow_method(n):
    wcss_list = []
    n = n + 1
    for i in range(1,n):
        wcss = KMeans(n_clusters=i, n_init='auto',).fit(transformed_data).inertia_
        wcss_list.append(wcss)

    sns.lineplot(x=np.arange(1,n), y = wcss_list)
    plt.grid(color = 'white', lw = .3)
    plt.show()
```

```python
elbow_method(31)
```
